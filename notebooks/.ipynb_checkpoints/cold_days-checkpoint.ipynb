```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_pickle('lga_min_temp_2000-2021.pkl')
```

# 2000 - 2004

```python
temps = data[data.Date < '2005-01-01']
fives = [5] * len(temps)
fifteens = [15] * len(temps)
below_five = len(temps[temps['Minimum Temperature'] < 5])
below_fifteen = len(temps[temps['Minimum Temperature'] < 15])
```

```python
plt.plot(range(len(temps)), temps['Minimum Temperature'])
plt.plot(fives)
plt.plot(fifteens)
plt.title('2000 - 2004... {} Days Below 5, {} Days Below 15'.format(below_five, below_fifteen))
plt.ylim([-5, 90])
```
